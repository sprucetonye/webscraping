```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bs4 as bs
import urllib.request

%matplotlib inline
```

```python
sauce = urllib.request.urlopen('https://www.brandonu.ca/scholarships/').read() 
```

```python
soup = bs.BeautifulSoup(sauce, 'lxml')
print(soup)
```

```python
print(soup.title)
```

```python
print(soup.title.string)
```

```python
print(soup.title.text)
```

```python
print(soup.p)
```

```python
print(soup.find_all('p')) 
```

```python
for paragraph in soup.find_all('p'):
    print(paragraph.string)
```

```python
for paragraph in soup.find_all('p'):
    print(paragraph.text)
```

```python
print(soup.get_text())
```

```python
for url in soup.find_all('a'):
    print(url)
```

```python
for url in soup.find_all('a'):
    print(url.get('href'))
```

```python
nav = soup.nav
print(nav)
```

```python
for url in nav.find_all('a'):
    print(url.get('href'))
```
